```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
m = 1
l = 1
D = 5
g = 9.8
```

```python
dt = 0.01
t_start = 0
t_end = 1
t_eval = np.arange(t_start, t_end, dt)
```

```python
q_0 = 0.1
omega_0 = 0
M_0 = 0
```

```python
def pendulum(t, y):
    q, omega, M = y
    
    domega = - (g/l) * np.cos(q) + M / (m * l**2)
    dq = omega
    dM = -D * omega

    return [dq, domega, dM]
```

```python
y0 = [q_0, omega_0, M_0]

sol = solve_ivp(pendulum, [t_start, t_end], y0, t_eval=t_eval)
```

```python
t = sol.t
q = sol.y[0]
omega = sol.y[1]
M = sol.y[2]
```

```python
plt.plot(t, q, label='q')
plt.plot(t, omega, label='omega')
plt.plot(t, M, label='M')
plt.title('Pendulum')
plt.xlabel('t')
plt.grid()
plt.legend()
plt.show()
```
